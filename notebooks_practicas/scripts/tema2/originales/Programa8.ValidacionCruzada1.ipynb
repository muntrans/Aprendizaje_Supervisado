```python
from sklearn.model_selection import KFold
```

```python
# Carga de datos.
X = ["a", "b", "c", "d", "e", "f", "g", "h", "i", "j"]
```

```python
# Validación cruzada.
kf = KFold(n_splits = 5, shuffle=True) # Shuffle permite aleatorizar las bolsas de test vez de hacer una lectura por orden -- KFold es una clase de scikit-learn.
  # n_splits = 5: Divide los datos en 5 partes
  # shuffle=True: Baraja aleatoriamente antes de dividir
bolsas = kf.split(X) # Genera los índices para cada fold --> esto devuelve un generador, no una lista
print(bolsas)
```

```python
# Mostrar ejemplos de cada iteración.
k = 1
# El bucle itera por cada fold y muestra qué índices van a train y cuáles a test
for train, test in bolsas:
    print("Iteracion", k, ":")
    print(" - Entrenamiento: %s" % (train))
    print(" - Test: %s" % (test))
    k = k + 1
```
